```python
import pandas as pd
import numpy as np 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
df
```

```python
df.isnull()
```

```python
df_1 = df.dropna()
df_1
```

```python
df_1.drop(columns=['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'high_blood_pressure', 
                     'platelets','serum_sodium', 'sex', 'smoking', 'time'], inplace=True)


df_1
```

```python
X = df_1[['ejection_fraction', 'serum_creatinine']].values
y = df_1[['DEATH_EVENT']]
```

Next, we need to split the data into training and testing sets using the train_test_split function from sklearn. We will use 70% of the data for training and 30% for testing.



```python
X_train,X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state = 42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
X_new = np.array([[20, 1.7], [40,6.2], [40, 1.9]])
y_pred = model.predict(X_new)
```

```python
y_pred
```

```python
model.score(X_test, y_test)
```
